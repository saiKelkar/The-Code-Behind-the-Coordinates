```python
import numpy as np
import sympy as sym
from IPython.display import display, Math
import matplotlib.pyplot as plt
```

```python
# Row vector and a Column vector
row_vec = np.array([4, 2])
col_vec = np.array([ [-2], [3] ])

print("Row Vector:", row_vec)
print("Column Vector:", col_vec)

# Render in LaTeX
display(Math(sym.latex(sym.simplify(row_vec))))
display(Math(sym.latex(sym.simplify(col_vec))))
```

```python
# Visualizing scalar multiplication
v = np.array([.5, 1])
s = [1, -.5, 2, .5]

for si in s:
    sv = si * v
    plt.plot([0, sv[0]], [0, sv[1]], 'o-', linewidth=3, label=f'$\\lambda$ = {si}')

plt.axis('square')
plt.axis([-3, 3, -3, 3])
plt.grid()
plt.legend()
plt.show()
```

```python
# Adding vectors
v1 = np.array([-1, 2])
v2 = np.array([1, 1])

# Both techniques yield the same result
v3a = v1 + v2
v3b = np.add(v1, v2)

print("Standard addition:", v3a)
print("NumPy addition:", v3b)
```

```python
# Dot Product

m_1 = np.arange(10, 20, 3)
m_2 = np.arange(5, 15, 3)

d_p = np.dot(m_1, m_2)
print("Dot Product:", d_p)
```

```python
# Matrix Multiplication
A = np.random.rand(3, 2) # 3 rows, 2 columns
B = np.random.rand(2, 4) # 2 rows, 4 columns

# Method 1: Using the @ operator
C = A @ B

# Method 2: Using np.dot
D = np.dot(A, B)

print("Matrix C Shape: ", np.shape(C))
print(C)

print("Matrix D Shape: ", np.shape(D))
print(D)
```
